### Random tensors
random tensors can create by using `torch.rand()`

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
random_tensor.shape
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(224,224,3))
random_image_size_tensor.shape , random_image_size_tensor.ndim
```

### Zeros and ones

```python
# Create zeros tensor
zero = torch.zeros(size=(3,4))
zero
```

```python
zero * random_tensor
```

```python
# Create ones tensor
ones = torch.ones(size=(3,4))
ones
```

```python
ones.dtype
```

### Creating a range of tensors and tesnors-like

```python
# Use torch.arange()
one_to_ten = torch.arange(1,10,2)
one_to_ten
```

```python
# Creating tesnors like
zeros_like = torch.zeros_like(one_to_ten)
zeros_like
```

### Tensor datatype

```python
# Float 32 tensor
float_32_tensor = torch.tensor([4,4.0])
float_32_tensor , float_32_tensor.dtype
```

```python
float_16_tensor = torch.tensor([4,4.0],dtype=torch.half)
float_16_tensor , float_16_tensor.dtype
```

```python
float_16_tensor_v2 = float_32_tensor.type(torch.half)
float_16_tensor_v2 , float_16_tensor_v2.dtype
```

```python
(float_16_tensor * float_32_tensor).dtype
```

### Getting information from tensors


```python
some_tensor = torch.rand(3,4)
some_tensor
```

```python
# Information of tensor
print("shape:",some_tensor.shape)
print("size:",some_tensor.size())
print("device:",some_tensor.device)
print("dim:",some_tensor.ndim)
print("dtype:",some_tensor.dtype)
```
