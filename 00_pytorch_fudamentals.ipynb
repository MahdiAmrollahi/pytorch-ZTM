### Reshaping, Stacking, Sqeezing tensors

```python
x = torch.arange(1,10)
x , x.shape
```

```python
# maltipication reshape size need to = shape
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
# change the view
x_view = x.view(1,9)
x_view , x_view.shape
```

```python
# changin x_view changes original x 
x_view[:,0]=5
x_view , x
```

```python
# stack
x_stack = torch.stack([x,x,x],dim=0)
x_stack
```

```python
x_stack = torch.stack([x,x,x],dim=1)
x_stack
```

```python
x_hstack = torch.hstack([x,x,x])
x_hstack
```

```python
x_vstack = torch.vstack([x,x,x])
x_vstack
```

```python
# Squeeze --> remove all single dim (1 dim)
x_reshaped
```

```python
x_reshaped.shape
```

```python
x_squeezed = x_reshaped.squeeze()
x_squeezed
```

```python
x_squeezed.shape
```

```python
# Unsqueeze --> add single dimension (1 dim)
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed
```

```python
x_unsqueezed.shape
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=1)
x_unsqueezed
```

```python
x_unsqueezed.shape
```
